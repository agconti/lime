```python
from pandas import read_csv
import datetime
```

```python
#http://www.netfonds.no/quotes/tradedump.php?paper=TSLA.O&csv_format=csv
url = 'http://www.netfonds.no/quotes/tradedump.php'
start_date = datetime.datetime.today().strftime('%Y%m%d')
ticker = "TSLA.O"
file_format = "csv"
uri= '{}?date={}&paper={}&csv_format={}'.format(url,
                                                start_date,
                                                ticker,
                                                file_format)
uri
```

```python
read_csv(uri)
```
